```python
import numpy as np
import pandas as pd
batch_num_image_idx_info_path = '/Users/sam/All-Program/App-DataSet/DeepFaceRecognition/data_models/batch_img_arr/batch_num_image_idx_info.csv'
person_name_image_num_info_path= '/Users/sam/All-Program/App-DataSet/DeepFaceRecognition/data_models/person_name_image_num_info.csv'
cv_prediction_analysis_fld_1_epch_1_path = '/Users/sam/All-Program/App-DataSet/DeepFaceRecognition/analysis/cv_pred_analysis/cv_prediction_analysis_fld_1_epch_1.csv'
```

```python
batch_num_image_idx_info = pd.read_csv(batch_num_image_idx_info_path)
batch_num_image_idx_info.head()
```

```python
batch_num_image_idx_info_in = batch_num_image_idx_info[batch_num_image_idx_info['batch_num'] == 10].reset_index().drop('level_0', axis=1)
print (batch_num_image_idx_info_in.shape)
batch_num_image_idx_info_in.head()
```

```python
cv_prediction_analysis_fld_1_epch_1 = pd.read_csv(cv_prediction_analysis_fld_1_epch_1_path)
print (cv_prediction_analysis_fld_1_epch_1.shape)
cv_prediction_analysis_fld_1_epch_1.head()
```

```python
p_name_batch_num_mrgd = person_name_image_num_info.merge(batch_num_image_idx_info_in, left_on=['index', 'image_label'], right_on=['index', 'image_label'], how='inner')
print (p_name_batch_num_mrgd.shape)
p_name_batch_num_mrgd.head()
```

```python
finalOPT = pd.concat([p_name_batch_num_mrgd, cv_prediction_analysis_fld_1_epch_1], axis=1, join_axes=[p_name_batch_num_mrgd.index])
```
